#Cleaning Data

##Deleting NaN and Unused Column

```python
df.drop(columns=["id","Unnamed: 32"],inplace=True)
df.isna().sum()
```

#Preprocessing

##Splitting X and y

```python
X = df.iloc[:,1:].values
X
```

```python
y = df.iloc[:,0].values
y
```

##Train Test Split

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

#Making Models

```python
models = []
labels = ["Logistic Regression","KNN","Decision Tree","Random Forrest","Naive Bayes","SVM","Kernel SVM"]
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.preprocessing import StandardScaler
logistic = LogisticRegression(random_state = 0)
models.append(logistic.fit(X_train, y_train))
```

```python
from sklearn.neighbors import KNeighborsClassifier
classifier = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p =2)
models.append(classifier.fit(X_train, y_train))
```

```python
from sklearn.tree import DecisionTreeClassifier
classifier = DecisionTreeClassifier(criterion = 'entropy',random_state=0)
models.append(classifier.fit(X_train,y_train))
```

```python
from sklearn.ensemble import RandomForestClassifier
classifier = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
models.append(classifier.fit(X_train, y_train))
```

```python
from sklearn.naive_bayes import GaussianNB
classifier = GaussianNB()
models.append(classifier.fit(X_train, y_train))
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'linear', random_state = 0)
models.append(classifier.fit(X_train, y_train))
```

```python
from sklearn.svm import SVC
classifier = SVC(kernel = 'rbf', random_state = 0)
models.append(classifier.fit(X_train, y_train))
```

```python
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score
from sklearn.model_selection import cross_val_score, StratifiedKFold
from sklearn.model_selection import cross_validate
accuracy = []
precision  = []
recall = []
f1= []
for i in range(len(models)) :

  cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)

  scoring = {
      'accuracy': make_scorer(accuracy_score),
      'precision': make_scorer(precision_score, average='weighted'),
      'recall': make_scorer(recall_score, average='weighted'),
      'f1': make_scorer(f1_score, average='weighted')
  }


  scores = cross_val_score(models[i], X, y, cv=cv, scoring='accuracy')


  results = cross_val_score(models[i], X, y, cv=cv, scoring='accuracy')


  multi_scores = cross_validate(models[i], X, y, cv=cv, scoring=scoring)

  accuracy.append(multi_scores['test_accuracy'].mean())
  precision.append(multi_scores['test_precision'].mean())
  recall.append(multi_scores['test_recall'].mean())
  f1.append(multi_scores['test_f1'].mean())
```

#Evaluating Models

```python
result_df = pd.DataFrame({
    "Algorithm":labels,
    "Accuracy":accuracy,
    "Precision":precision,
    "Recall":recall,
    "F1 Score":f1
  }
)
result_df.sort_values(by='Accuracy', ascending=False)
```
