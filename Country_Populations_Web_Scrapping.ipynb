#Importing BeautifulSoup 

```python
import requests
from bs4 import BeautifulSoup
```

#Getting URL 

```python
URL = "https://www.scrapethissite.com/pages/simple/"
```

#Requesting

```python
page = requests.get(URL)
```

#Using BeautifulSoup

```python
soup = BeautifulSoup(page.content, "html.parser")
```

```python
countries = soup.find(id = "countries")
```

```python
print(countries.prettify)
```

#Getting Country Names and Population 

```python
countries_name = countries.find_all("h3",class_="country-name")
countries_pop = countries.find_all("span",class_="country-population")
```

```python
countries_pop_data = []
for element in countries_pop: 
  countries_pop_data.append(int(element.text))
```

```python
countries_name_data = []
for element in countries_name:
  countries_name_data.append(element.text.replace("\n","").replace("                        ","").replace("    ",""))
```

#Using Python Pandas

```python
import pandas as pd
```

```python
data = { 
    "Country Name":countries_name_data,
    "Population":countries_pop_data
}
```

```python
df = pd.DataFrame(data)
```
