## Exploratory Data Analysis (EDA) 

```python
fig, axs = plt.subplots(3,4,figsize=(18,12))
inc = 0
for i in range(3):
    for j in range(4):
        sns.histplot(data=df,x=df.columns[j+inc],kde=True,ax=axs[i,j],color="#003f5c")
    inc += 4 
```

```python
fig, ax = plt.subplots(figsize=(16,10))
sns.heatmap(df.corr(),annot=True, linewidths=0.5)
```

```python
fig, axs = plt.subplots(2,2,figsize=(12,9))
sns.scatterplot(data=df,x="fixed acidity", y="density",ax=axs[0,0],color="#003f5c")
sns.scatterplot(data=df,x="fixed acidity", y="citric acid",ax=axs[0,1],color="#003f5c")
sns.scatterplot(data=df,x="free sulfur dioxide", y="total sulfur dioxide",ax=axs[1,0],color="#003f5c")
```

```python
fig, axs = plt.subplots(2,2,figsize=(12,9))
sns.scatterplot(data=df,x="pH", y="fixed acidity",ax=axs[0,0],color="#003f5c")
sns.scatterplot(data=df,x="pH", y="citric acid",ax=axs[0,1],color="#003f5c")
sns.scatterplot(data=df,x="citric acid", y="volatile acidity",ax=axs[1,0],color="#003f5c")
```

```python
label = ["5","6","7","4","8","3"]
explode = [0.2,0,0,0,0,0]
plt.pie(df["quality"].value_counts(),labels=label,shadow=True,startangle=90)
plt.title("Wine Quality")
plt.legend()
```

```python
fig, axs= plt.subplots(1,3,figsize=(12,3))

sns.barplot(data=df,x="quality",y="alcohol",color="#003f5c",ax=axs[0])
sns.barplot(data=df,x="quality",y="sulphates",color="#bc5090",ax=axs[1])
sns.barplot(data=df,x="quality",y="citric acid",color="#ffa600",ax=axs[2])
```
